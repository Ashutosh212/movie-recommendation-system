As above, we can see all the features expect `id` is object(i.e they are string), we somehow have to extract important imformation from them by writing some helper function.

We will see all the features column string one-by-one and try to extract the important name or attributes from them by ignoring all metadata.

1. keywords

```python
df_final.iloc[0]['keywords']
```

we got the list of dictionary in which we got the tag of movie, but this is a string. We have to write 
function to extract name of each keywords.

```python
import json

def extract_keywords(s):
    # Convert to Python list of dicts
    data = json.loads(s)
    names = [item['name'] for item in data]
    return names

df_final['keywords'] = df_final['keywords'].apply(extract_keywords)
```

3. production_companies

```python
df_final.iloc[5]['production_companies']
```

```python
def extract_companies(data):
    data = json.loads(data)
    return [item['name'] for item in data]

df_final['production_companies'] = df_final['production_companies'].apply(extract_companies)
```

```python
df_final['production_companies']
```

Let's count the production companies which are produced most films, these are only needed to be stored. 

```python
companies_count = {}

for i in range(len(df_final)):
    comp_list = df_final.iloc[i]['production_companies']
    for comp in comp_list:
        count = companies_count.get(comp, 0)
        count += 1
        companies_count[comp] = count
```

```python
companies_count = dict(sorted(companies_count.items(), key= lambda x : x[1], reverse=True))
```

```python
dict(companies_count)
```

I am only going to consider which have atleast 50 movies in the dataset and remove all 

```python
def clean_production_companies(data):
    new_list = []
    for comp in data:
        if companies_count[comp] > 50:
            new_list.append(comp)
    return new_list

df_final['production_companies'] = df_final['production_companies'].apply(clean_production_companies)
df_final['production_companies']
```

```python
def get_cast_name(data):
    name_list = []
    data = json.loads(data)
    return [item['name'] for item in data[:3]]

df_final['cast'] = df_final['cast'].apply(get_cast_name)
df_final['cast']
```

5. Crew

```python
df_final['crew'][0]
```

```python
def get_director_name(data):
    data = json.loads(data)
    for item in data:
        if item['job'] == "Director":
            return item['name']

df_final['crew'] = df_final['crew'].apply(get_director_name)
df_final['crew']
```

6. genres

```python
df_final['genres'][0]
```

```python
def get_genres(data):
    data = json.loads(data)
    genre = []
    for item in data:
        genre.append(item['name'])
    return genre

df_final['genres'] = df_final['genres'].apply(get_genres)
df_final['genres']
```
