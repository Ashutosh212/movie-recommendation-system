```python
df_final.iloc[0]['keywords']
```

we got the list of dictionary in which we got the tag of movie, but this is a string. We have to write 
function to extract name of each keywords.

```python
import json

def extract_keywords(s):
    # Convert to Python list of dicts
    data = json.loads(s)
    names = [item['name'].lower() for item in data]
    return names

df_final['keywords'] = df_final['keywords'].apply(extract_keywords)
```

```python
def extract_companies(data):
    data = json.loads(data)
    return [item['name'].lower() for item in data]

df_final['production_companies'] = df_final['production_companies'].apply(extract_companies)
```

```python
df_final['production_companies']
```

Let's count the production companies which are produced most films, these are only needed to be stored. 

```python
companies_count = {}

for i in range(len(df_final)):
    comp_list = df_final.iloc[i]['production_companies']
    for comp in comp_list:
        count = companies_count.get(comp, 0)
        count += 1
        companies_count[comp] = count
```

```python
companies_count = dict(sorted(companies_count.items(), key= lambda x : x[1], reverse=True))
```

```python
dict(companies_count)
```

I am only going to consider which have atleast 50 movies in the dataset and remove all 

```python
def clean_production_companies(data):
    new_list = []
    for comp in data:
        if companies_count[comp] > 50:
            new_list.append(comp.lower())
    return new_list

df_final['production_companies'] = df_final['production_companies'].apply(clean_production_companies)
df_final['production_companies']
```

## Sematic Search

For sematic search, we don't need furthur processing like stemming and remove the white space. We will do it later when we do hybrid search. 

```python
df_final.head()
```

```python
df_final.info()
```

Concatenation of all column except id and title to make meaning embedding.


```python
def build_movie_text(row):
    parts = [
        f"Genres: {" ".join(row['genres'])}",
        f"Keywords: {', '.join(row['keywords'])}",
        f"Overview: {row['overview']}",
        f"Production Companies: {', '.join(row['production_companies'])}",
        f"Cast: {', '.join(row['cast'])}",
        f"Director: {row['crew']}",
    ]
    return " ".join(parts)
```

```python
build_movie_text(df_final.iloc[5])
```

```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI()

# Example from documentation
# response = client.embeddings.create(
#     input="Your text string goes here",
#     model="text-embedding-3-small",
#     dimensions=256
# )

# print(response.data[0].embedding)
```

```python
def get_embedding(text: str):
    response = client.embeddings.create(
        input=text,
        model="text-embedding-3-small",
        dimensions=256
    )
    return response.data[0].embedding
```

```python
all_embeddings = []  


for i in range(len(df_final[:100])):
    context = build_movie_text(df_final.iloc[i])
    # print(context)
    embd = get_embedding(context)
    all_embeddings.append(embd)
    

embedding_matrix = np.array(all_embeddings, dtype=np.float32)  # shape: (num_movies, 1536)
```

```python
np.save("movie_embeddings.npy", embedding_matrix)
```

```python
embedding_matrix = np.load("movie_embeddings.npy")
embedding_matrix.shape
```

```python
norms = np.linalg.norm(embedding_matrix, axis=1, keepdims=True)
normalized_matrix = embedding_matrix / norms
```

```python
query = get_embedding("Romantic movie")
query = query / np.linalg.norm(query)

# Compute cosine similarities
scores = np.dot(normalized_matrix, query)

# Get top-k
top_k_idx = np.argsort(-scores)[:5]  # descending sort

# Print results
for i in top_k_idx:
    print("------")
    print(f"Title: {df_final.iloc[i]['title']}")
    print(f"Overview: {df_final.iloc[i]['overview']}")
    print("------")
```

## Text Search

### Converting to lowercase

It is good practise in NLP to convert all the text to lowercase before doing any embedding task for consistency.

```python
def to_lower(data):
    if isinstance(data, str):
        return data.lower()
    elif isinstance(data, list):
        return [item.lower() for item in data]


df_final['genres'] = df_final['genres'].apply(to_lower)
df_final['keywords'] = df_final['keywords'].apply(to_lower)
df_final['overview'] = df_final['overview'].apply(to_lower)
df_final['production_companies'] = df_final['production_companies'].apply(to_lower)
df_final['cast'] = df_final['cast'].apply(to_lower)
df_final['crew'] = df_final['crew'].apply(to_lower)
```
